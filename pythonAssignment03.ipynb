In a dice game, a die is rolled $3$ times, and the total number of eyesfrom the three rolls is counted. The random variable X represents the sum of the three rolls. The die is assumed to be fair.

```python
import numpy as np
import matplotlib.pyplot as plt
dice = {1,2,3,4,5,6}
rolls = 0
for dice1 in dice:
    for dice2 in dice:
        for dice3 in dice:
            rolls+=1
```

### 1 How many possibilities are there for $X$ to be $3,4,5,...,18$?

```python

occurance = 0

target = 3

for target in range(3,19):
    occurance = 0
    for dice1 in dice:
        for dice2 in dice:
            for dice3 in dice:
                if dice1+dice2+dice3==target:
                    occurance+=1
    print(f'X={target}: {occurance}')


```

### 2 Plot the PMF and CDF for $X$

```python
# Calculate the probabilities for each possible value of X
probabilities = {}
for target in range(3, 19):
    occurance = 0
    for dice1 in dice:
        for dice2 in dice:
            for dice3 in dice:
                if dice1 + dice2 + dice3 == target:
                    occurance += 1
    probabilities[target] = occurance / rolls

# Extract the values and probabilities
values = list(probabilities.keys())
pmf = list(probabilities.values())

# Calculate the CDF
cdf = np.cumsum(pmf)

# Plot the PMF
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.stem(values, pmf)
plt.title('PMF of X')
plt.xlabel('Sum of rolls (X)')
plt.ylabel('Probability')

# Plot the CDF
plt.subplot(1, 2, 2)
plt.step(values, cdf, where='mid')
plt.title('CDF of X')
plt.xlabel('Sum of rolls (X)')
plt.ylabel('Cumulative Probability')


plt.show()
```

### 3 What is the probability for $9\leq X \leq 15$?

```python
occurance = 0
for target in range(9,16):
    for dice1 in dice:
        for dice2 in dice:
            for dice3 in dice:
                if dice1 + dice2 + dice3 == target:                    
                    occurance += 1

print(occurance/rolls)
```

```python
res = {}
dice = {1,2,3,4,5,6}
for target in range(3,19):
    count = 0
    for dice1 in dice:
        for dice2 in dice:
            for dice3 in dice:
                if dice1+dice2+dice3==target:
                    count+=1
    res[target]=count



totalSum = 0
for j in res:
    totalSum += j*res[j]

print(totalSum / rolls)
```
